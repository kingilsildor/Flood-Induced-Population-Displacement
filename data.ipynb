```python
import pandas as pd


def filter_dataframe(df, location):
    location_cols = [f"{location} sim", f"{location} data", f"{location} error"]

    valid_location_cols = [col for col in location_cols if col in df.columns]

    if "Total error" in df.columns:
        extra_cols = df.columns[df.columns.get_loc("Total error") :].tolist()
    else:
        extra_cols = []

    columns_to_keep = ["Date"] + valid_location_cols + extra_cols

    return df[columns_to_keep]


df_filtered = filter_dataframe(output, "Camp_5")
df_filtered
```
